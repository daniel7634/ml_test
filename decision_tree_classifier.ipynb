```python
from sklearn import tree
from sklearn import datasets
from matplotlib import pyplot
```

```python
iris = datasets.load_iris()
x = iris.data
y = iris.target
```

```python
clf = tree.DecisionTreeClassifier(criterion='entropy').fit(x, y)
```

```python
def show_tree_figure(clf, feature_names, class_names):
  pyplot.figure(figsize=(10, 5))
  tree.plot_tree(clf, filled=True, feature_names=feature_names, class_names=class_names)
  pyplot.show()

show_tree_figure(clf, iris.feature_names, iris.target_names)
```

# 拆分 train data 和 test data

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)

clf = tree.DecisionTreeClassifier(criterion='entropy').fit(x_train, y_train)
show_tree_figure(clf, iris.feature_names, iris.target_names)

print(f'train accuracy: ${clf.score(x_train, y_train)}')

y_pred = clf.predict(x_test)
print(f'predict: ${y_pred}')

accuracy = clf.score(x_test, y_test)
print(f'test accuracy: ${accuracy}')
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)

clf = tree.DecisionTreeClassifier(criterion='gini').fit(x_train, y_train)
show_tree_figure(clf, iris.feature_names, iris.target_names)

print(f'train accuracy: ${clf.score(x_train, y_train)}')

y_pred = clf.predict(x_test)
print(f'predict: ${y_pred}')

accuracy = clf.score(x_test, y_test)
print(f'test accuracy: ${accuracy}')
```

# Overfitting 調整

```python
clf = tree.DecisionTreeClassifier(criterion='entropy', max_depth=3).fit(x_train, y_train)
show_tree_figure(clf, iris.feature_names, iris.target_names)

print(f'train accuracy: ${clf.score(x_train, y_train)}')

y_pred = clf.predict(x_test)
print(f'predict: ${y_pred}')

accuracy = clf.score(x_test, y_test)
print(f'test accuracy: ${accuracy}')
```
